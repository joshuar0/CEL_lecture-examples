```python
import numpy as np
from scipy.sparse import coo_matrix
```

```python
def read_alist_file(filename: str):
    try:
        with open(filename, "rt+") as f:
            t = f.readline().split(" ")
            N = int(t[0])
            M = int(t[1])

            t = f.readline().split(" ")
            max_sum_dv = int(t[0])
            max_sum_dc = int(t[1])

            sum_dv = np.fromstring(f.readline(), dtype=int, sep=" ")
            sum_dc = np.fromstring(f.readline(), dtype=int, sep=" ")

            row_idx = []
            for _ in range(N):
                row_idx.extend(map(int, f.readline().split()))

            f.close()
        
        col_idx = np.repeat(np.arange(1, N+1), max_sum_dv)

        row_idx = row_idx[row_idx>0] - 1
        col_idx = col_idx[row_idx>0] - 1

        return coo_matrix((np.ones(len(row_idx)), (row_idx, col_idx)), shape=(M,N))
        
    except Exception as e:
        print(e)
```
